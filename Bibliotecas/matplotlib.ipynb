```python
import pandas as pd
import matplotlib.pyplot as plt
import time
import datetime 
import unittest
import numpy as np
```

Visualização de dados com Matplolib

```python
np.random.seed(7)

y = np.random.randint(low=1, high=1500, size=10)

y
```

```python
plt.plot(y)
```

```python
# insere a primeira linha no plot

plt.plot(y, color='#749187', marker='o', ms=5, mec='k', markerfacecolor='w', ls='-.')

# insere a segunda linha no plot
plt.plot(y*2, marker='+', color = 'k', ms = 5)

# rótulos

plt.xlabel('Eixo X', color = 'red',size = 10)
plt.ylabel('Eixo Y', color = 'red',size = 10)
plt.title('Gráfico Matplolib', loc= 'center', color = 'red')

# gridlines

plt.grid(axis = 'y', color = 'gray',
         linestyle = '--', linewidth = 1,
         alpha = 0.8)
plt.show()
```

```python
np.random.seed(6)
x = np.arange(1, 11)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500,10)
y3 = np.random.randint(200, 600, 10)

plt.figure(figsize = (15, 5))
plt.suptitle('Figure', fontsize = 15)

plt.subplot(1, 3, 1)
plt.plot(x, y1, color = 'k')  # cor preta
plt.title('Gráfico 1', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')


plt.subplot(1, 3, 2)
plt.plot(x, y1, color = 'r') # cor vermelha
plt.title('Gráfico 2', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 3)
plt.plot(x, y1, color = 'g') # cor verde
plt.title('Gráfico 3', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.tight_layout(pad = 4) # espaçamento entre os gráficos

plt.show()
```

```python
# Outra forma de criar os gráficos com o matplotlib

fig, ax = plt.subplots(1,3, figsize = (15,5))
fig.suptitle('Projeto')
ax[0].plot(x, y1, color = 'black')
ax[1].plot(x, y2, color = 'red')
ax[2].plot(x, y3, color = 'green')

for i in range(3):
    ax[i].set(title = f'Gráfico {i+1}', xlabel = 'Eixo X', ylabel = 'Eixo Y')
```
